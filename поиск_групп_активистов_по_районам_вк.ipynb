```python
# удаляем строку всего

only_districts = districts['Район']
only_districts = only_districts.drop(index = 125)
```

```python
# колонка для числа групп сообществ
only_districts['group_number'] = ''
```

```python


for i in range(len(only_districts)):
    print(only_districts.iloc[i,0])
    results = api.groups.search(q = only_districts.iloc[i,0], city_id = 1,v='5.58',count=1000)
    
    counter = 0
    id_list = []
    for j in range(len(results['items'])):
        buffer = results['items'][j]['name']
        if (only_districts.iloc[i,0][:-2] in buffer or only_districts.iloc[i,0][:-2].upper() in buffer)  and ('амоуправление' in buffer or 'ольщики' in buffer or 'против'in buffer or 'протест'in buffer or 'ащита'in buffer or 'нициативн'in buffer or 'еноваци'in buffer or 'пикет'in buffer or 'итинг'in buffer or 'ТСЖ'in buffer or 'мой район' in buffer or 'АМОУПРАВЛЕНИЕ' in buffer or 'ПРОТИВ'in buffer or 'ПРОТЕСТ'in buffer or 'АЩИТА'in buffer or 'ИНИЦИАТИВН'in buffer or 'ЕНОВАЦИ'in buffer  or 'ПИКЕТ'in buffer or 'ИТИНГ'in buffer or 'ТСЖ'in buffer or 'МОЙ РАЙОН' in buffer or 'ОЛЬЩИКИ' in buffer):
            print(buffer)
            counter += 1
            id_list.append(results['items'][j]['id'])
            
            
    only_districts.iloc[i,1] =    counter
    
    total_number = 0
    for m in id_list:
        try:
            people = api.groups.getMembers(group_id = str(m),v='5.58')
            total_number = total_number + people['count']
        except:
            continue
    only_districts.iloc[i,2] =    total_number
    print(only_districts.iloc[i,2], "    ",only_districts.iloc[i,1])
    
   

    
```
